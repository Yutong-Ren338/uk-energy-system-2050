```python
# Array of CAPEX
CAPEX_arr = np.arange(1500,8800,100)
```

```python
# Calculate Annualised cost for DAC
currency = 1/1.27 #USD/GBP=0.79
discount_rate = 0.05
opex = 0.02
life = 20
Annual_cost = CAPEX_arr * currency * 10**6 * (discount_rate/(1-(1/(1+discount_rate))**life)+opex)
```
