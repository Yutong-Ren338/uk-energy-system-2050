# System Boundary Schematic

This notebook produces a single-node system boundary diagram that matches the report paragraph.
It is a conceptual schematic (no data required).


```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyArrowPatch
import src.matplotlib_style  # noqa: F401

def add_box(ax, xy, w, h, text, fc="#F3F6FA", ec="#1F2A44"):
    rect = Rectangle(xy, w, h, linewidth=1.5, edgecolor=ec, facecolor=fc)
    ax.add_patch(rect)
    ax.text(
        xy[0] + w / 2,
        xy[1] + h / 2,
        text,
        ha="center",
        va="center",
        fontsize=9,
    )
    return rect

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 7)
ax.axis("off")

# Main node
add_box(
    ax,
    (3.0, 2.6),
    4.0,
    1.8,
    "Single-node UK Electricity System\n(aggregated demand/supply)",
    fc="#E8F0FE",
)

# Generation portfolio
add_box(
    ax,
    (0.5, 3.4),
    2.2,
    2.6,
    "Generation\nSolar PV\nOnshore wind\nOffshore wind\nNuclear\nGas-CCS",
)

# Demand
add_box(
    ax,
    (7.3, 3.6),
    2.2,
    1.4,
    "Demand\n(2050 projection\nSeventh Carbon Budget)",
)

# Imports (optional)
add_box(
    ax,
    (0.5, 6.1),
    2.2,
    0.7,
    "Interconnector imports\n(optional)",
    fc="#FFF4E6",
)

# Curtailment
add_box(
    ax,
    (7.3, 6.1),
    2.2,
    0.7,
    "Curtailment\n(surplus)",
    fc="#FDECEF",
)

# Flexibility options
add_box(
    ax,
    (1.4, 0.5),
    2.4,
    1.2,
    "Electricity storage\n(medium-term)",
)
add_box(
    ax,
    (4.0, 0.5),
    2.4,
    1.2,
    "Hydrogen system\nElectrolysis\nStorage\nH2 generation",
)
add_box(
    ax,
    (6.6, 0.5),
    2.4,
    1.2,
    "DAC\n(uses electricity)",
)

def arrow(start, end, style="-|>", ls="-"):
    ax.add_patch(
        FancyArrowPatch(
            start,
            end,
            arrowstyle=style,
            mutation_scale=12,
            linewidth=1.2,
            linestyle=ls,
            color="#1F2A44",
        )
    )

# Flows
arrow((2.7, 4.6), (3.0, 3.5))  # generation to node
arrow((7.0, 3.5), (7.3, 4.3))  # node to demand
arrow((2.7, 6.45), (3.1, 4.4), ls="--")  # imports to node
arrow((7.0, 4.8), (7.3, 6.1), ls="--")  # node to curtailment

# Flexibility arrows (bidirectional where appropriate)
arrow((3.8, 2.6), (2.6, 1.7))
arrow((2.6, 1.7), (3.8, 2.6), style="-|>")
arrow((5.0, 2.6), (5.2, 1.7))
arrow((5.2, 1.7), (5.0, 2.6), style="-|>")
arrow((6.8, 2.6), (7.4, 1.7))

# Loss factor note
ax.text(
    7.9,
    3.0,
    "T&D losses\n(scalar factor)",
    ha="center",
    va="center",
    fontsize=8,
    color="#4A5568",
)

# Resolution note
ax.text(
    5.0,
    0.1,
    "Daily resolution; energy in TWh/day (GW x 24)",
    ha="center",
    va="bottom",
    fontsize=9,
    color="#4A5568",
)

fig.tight_layout()
plt.show()
```
